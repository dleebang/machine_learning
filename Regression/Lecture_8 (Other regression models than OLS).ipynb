## Other regression models

The standard linear model is to use ordinary least squares to minimize the Residual Sum of Squares (RSS) from predicted values to the mean regression line

So, why look for alternatives?
1. You will have different prediction accuracies

Accuracy of ordinary least squares method is usually good when you have an approximately linear relationship between response and predictor variables, and when you have a lot of observations to regress. Particularly if you have to look at the balance of number of observations vs. number of predictors. If you reduce the number of predictor variables or shrink the predicted values towards 0, you can substantially reduce the variance of your model

2. You can have different model interpretability depending on the model

If we reduce the number of predictor variables we make interpretability of the model easier because you make it simpler. You can evaluate to drop or not a predictor variable looking at its beta coefficient. If it shows no relationship (low coefficient and p-value not siginficant), then that predictor variable has no impact on the dependent variable and can be dropped.

#### We will discuss two types of methods:

1. Subset selection
    
    Instead of training model on all variables, we will use a subset of available ones
    
    
2. Shrinkage method
    
    We will try to regularize i.e. shrink the coefficient of some variables toward zero.

## Subsection selection techniques

This approach involves identifying a **subset of p predictors** that we believe to be related to the response variable

There are three techniques for subsetting:

1. Best subset selection

2. Forward step-wise selection

3. Backward step-wise selection

## Shrinkage method (Regularization)

This approach involes fitting a model involving **all p predictors**. However, the estimated coefficients are shrunken towards zero relative to the least square estimates (betas). 

There are two techniques of Shrinkae method:

1. Ridge regression
2. Lasso 


**Ridge regression**

In ridge we try to shrink the coefficients of variable towards zero by adding shrinkage penalty of the residual sum of squares



Because of this penalty, Ridge regression varies with scale of independent variable, therefore, we need to standardize the values of these variables. The penalty also includes a $\lambda$ term that is called as the tuning-parameter which is used to control the relative impact of the RSS and the shrinkage penalty. 

We shrink all the estimates except the $\beta_{0}$ beause we do not want to shrink the intercept.

As the value of lambda parameters is increased, the model becomes even less flexible and therefore the model bias increases, but the variance is decreased. So the job is here is to find the lambda that minimizes the sum of bias and variance errors.



**Lasso regression**

In lasso we try to shrink the coefficients of variable towards zero by adding shrinkage penalty to the residual sum of squares

The shrinkage penalty is similar to the ridge one, except that it uses the absolute values of coefficient betas



In lasso, for sufficiently large value of $\lambda$, several coefficients betas will actually be forced to zero, resulting in predictor selection cutting off those ones that results in 0. 

The resulting model of lasso is usually more **interpretable** than ridge regression. However there is no dominance of one over another in terms of accuracy of the predictions.
