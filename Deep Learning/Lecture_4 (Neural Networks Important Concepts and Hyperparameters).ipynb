### Activation function

* **Why do we use activation functions after getting a value of z of multiplications of inputs*weights + bias for each perceptron to get an ouput?**

The answer is to put special boundary conditions on the ouput such as 0 and 1 limits for classification problems. When you have a regression problem, it is ok if you have only the results of of the perceptron calculation. Activation function is also used to capture non linearity of the data and find complex patterns of relationships.


* **What are the different types of activation functions?**

Earlier we discussed two types of activation functions, such as the step (0 and 1 threshold values) and sigmoide functions (continuous s-shaped curve with 0 and 1 boundaries). But as per the graph below, there are others as well: 



Other functions:
    * **hyperbolic tangent activation (tanh)**, the graph is similar to sigmoid, but has different boundaries (lower of -1 to higher 1). It is centered at 0 and thus can have better convergence than sigmoid
    * **rectified linear unit (ReLU)**. It is more used in regression neural networks. Until values of 0, the z will continue to be 0, but if greater, the activation ouput will be the same as of the input z. So the lower boundary is 0 with no upper boundary. This activation function can also capture non-linearity of data in hidden (intermediate) layers. However, it is hardly used in the ouput layer, because for classification problems you do not have boundaries to get 0 or 1 outputs.

Answer is yes. Some activation functions is preferred over others depending on the layer. 



* **What is multi class classification? Is there any specific activation function for this?**



Binary classification = Yes or No problem or 0 or 1

More than two classes like "shirts" "trousers" or "socks" it is a multiclass classification and the ouput cannot be 0 or 1. So, for multiclass classification we use an activation function called **softmax**. This activation function is similar to sigmoid, but has an additional step.

Softmax activation function: 

We have one neuron for each class to be predicted. 

For each class we keep one output neuron (final layer) all with sigmoid activation functions to find the probability of prediction of that neuron class. 

All the outputs of the ouput layer go into and additional softmax layer where each ouput is divided by the total sum to bring the total probability to one. 





## Gradient descent

* **What is the difference between Gradient Descent and Stochastic Gradient Descent?**

    * **Stochastic gradient descent**: Single training record (observations), forward and backward propagation. This is what we learned in the previous lecture of gradient descent. Stochastic gradient descent updat weights and biases rapidly, but it finds difficult converging. 

    * **Gradient descent**: Full training set, forward and backward propagation. Gradient descent is slower, because in each step, it has to go through the entire training set to find optimal weights and bias. But it converges very well. 

    * **Mini batch gradient descent**: small batch of the training set, forwad and backward propagation


* **What is the Vanishing gradients?**

    * It is a situation where a deep multilayer feed-foward network is unable to propagate useful gradient information to find optimal values of weights and bias from the output end of the model **back to the layers near the input end of the model**. The results is the general inability of models with many number of layers to learn on a given dataset, or for models with many layers to prematurely converge to a poor solution.

## Epoch

What is an epoch?

Epoch is one entire cycle of gradient descent method through the full training data

Note that it is different from iteration

For example: suppose we have 1000 training records (obs.), if we are apply the stochastic gradient descent algorithm, one record is punt at a time, then 1000 iterations witihn one epoch. Contrarily, if we use gradient descent, the algorithm will pass through the entire training set to find the optimal weights and bias. 

So, if we enter 1000 records training set 2 times, then **Epoch is 2**
