### Perceptron

#### Concept
Perceptron is a binary classifier which receives a layer of inputs $x_{1}, x_{2},...,x_{m}$ and each input gets a weight $w_{1}, w_{2},...,w_{m}$ to classify a binary ouput, either 0 or 1 if a given threshold is met. The condition to evaluate if the threshold is met is the actviation function that is given as $\sum_{j}w_{j}x_{j} < or > threshold$. Therefore, 0 output will be choosen if the thershold is not met, and 1 otherwise.




#### Removing binary restriction
We will add a bias term into the equation to compare with threshold.

Output = 0 if $\sum_{j}w_{j}x_{j} + bias < 0$ or 1 if $ \sum_{j}w_{j}x_{j} + bias >= 0 $

An **activation function** is the non-linear transformation that defines the output given a combination of inputs

Another activation functions can also be used such as the **sigmoid function** (also called as logistic function). It is a smooth shaped curve that has a minimum of 0 and minuns infinity, and maximum of 1 plus infinity. Contrary to step activation function, this function rises smoothly and gradually. 



The sigmoid is better because it is less sensitive to individual observation because small changes in weights of inputs can completely flip the output in the step function if you face a problem of misclassification and have to change weights. With sigmoid function, as the change is gradual from 0 to 1, it is easier to control the behavior of the algorithm.

Artificial neuron with sigmoid activation is called sigmoid or logistic neuron. The sigmoid function equation is: 

$$\sigma(z) \equiv \frac{1} {1 + e^{-z}}$$

Then, the ouput of the logistic neuron will be given as:

$$\sigma(z) \equiv \frac{1} {1 + exp(-\sum_{j}w_{j}x_{j} - bias)}$$

This value will always lie between 0 and 1 and will have a sigmoid and continuous shape. So this sigmoid equation will replace that conditions of step activation function of 0 and 1 if a given thershold is met

The activation function is important because when we haven't this function, the weights and biases simply result in a linear transformation.
