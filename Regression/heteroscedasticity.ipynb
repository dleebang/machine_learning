### Heteroscedasticity

This phenomenon occurs when $Var(\epsilon_{i}) = \sigma^{2}$ is heterogeneous troughout the distribution of points in the relationship of residual values and fitted (Y) values. The variance of error term is then independent of values of Y and **Non constant** (left graph below) and the boundaries of the relationship increase or decrease.

The solution of heteroscedasticity is to scale down Y variable by using log(y) or sqrt(y) instead
