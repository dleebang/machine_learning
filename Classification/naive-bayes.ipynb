## Naive Bayes

The Naive Bayes assumes total independency of the predictor variables. 

Naive Bayes is the calculation of posterior probabilities, that is, the probability of an event ocurring giving another event has ocurred. Therefore, the naive Bayes is the direct application of the calculation of the probabilities of observations ocurring. 

The Naive Bayes assumes normal distribution functions of the conditional probabilities given the predictor variabbles. 

The Naive Bayes approach is similar to logistic regression prediction mathematically, and the curve for estimates conditional probabilities also looks similar.



In the formula above, we calculate the probability of Y = 1 given the distribution of the predictor variable f1 divided by the same term (Y = 1 given the distribution of the predictor variable f1) + (Y = 0 (whith is 1-Pr(Y=1) given the distribution of the predictor variable f2). 

The Naive Bayes algorithm can work well in calculating the probabilities of categorical variables when you have a few number of predictors. But when the X gets high-dimensional (too many predictors) and we will not have many information about the distribution functions. In these cases, Naive Bayes will become practically impossible to implement.
