```python
#Outlier treatment
uv = np.percentile(df.n_hot_rooms,[99])[0] #find the 99th percentile
```

```python
df[(df.n_hot_rooms > uv)] #identify outliers above the 99th percentile
```

```python
df.n_hot_rooms[(df.n_hot_rooms > 3 * uv)] = 3 * uv #treat outliers with capping method
```

```python
lv = np.percentile(df.rainfall,[1])[0] #find the 1st percentile
```

```python
lv
```

```python
df[(df.rainfall < lv)] #identify outliers below the 1st percentile
```

```python
df.rainfall[(df.rainfall < 0.3 * lv)] = 0.3 * lv #treat below outliers with flooring method 
```

```python
df.describe() #the mean value is now closer to the median value for n_hot_rooms and rainfall
```

```python
#Missing values treatment
df.info() #identify missing values in counting
```

```python
df.n_hos_beds = df.n_hos_beds.fillna(np.mean(df.n_hos_beds))
```

## Logistic regression

#### Simple logistic regression with one single predictor

```python
#create X and Y variables
X = df[["price"]] # predictor variable is price
X
```

```python
y = df[["Sold"]]
y
```

```python
#import sklearn function
from sklearn.linear_model import LogisticRegression
```

```python
lrs = LogisticRegression() #create classfication object
```

```python
lrs.fit(X, y) #(independent, dependent)
```

```python
print(lrs.intercept_, lrs.coef_) # beta_0 (intercept) and beta_1 (coef) of the logist. regression model
```

## Logistic with multiple predictors 

```python
df.head()
```

```python
#create new variables
X = df.loc[:, df.columns != "Sold"]
X.head()
```

```python
y = df["Sold"]
y.head()
```

```python
mlrs = LogisticRegression()
mlrs.fit(X, y)
```

```python
print(mlrs.intercept_, mlrs.coef_) #beta 0 and betas 1, 2, ..., n
```

### Prediction and confusion matrix

```python
#predict probabilities of classes and classes itself with the coefficients of the model
mlrs.predict_proba(X) #get the probabilities of each class; the ouput is two columns, first with the probs of being 0, second
#with the probs of being 1
```

```python
y_pred = mlrs.predict(X) #predict the classes itself with the given probabilities; the decision rule by default is 0.5
#this means that if probs is >0.5 you predict 1, otherwise, you predict 0
y_pred
```

```python
y_pred_03 = (mlrs.predict_proba(X)[:,1] >= 0.3) #set an abritrary decision rule of 0.3
y_pred_03 #the output will be an array of trues and falses given the condition of the decision rule
#True = 1 (Sold)
#False = 0 (Not sold)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y, y_pred) #compare the predicted outcomes with true outcomes
#[true positives, false positives]
#[false negatives, true negatives]
```

```python
#create confusion matrix for the decision rule of 0.3
confusion_matrix(y, y_pred_03) 
```

#### Performance metrics



* precision = specificity (predicted positive values or true negative rate)
* recall = sensitivity (true positive rate)

* ROC curve

```python
from sklearn.metrics import precision_score, recall_score, classification_report, roc_auc_score, plot_roc_curve
```

```python
precision_score(y, y_pred) #precision = specificity (true negative rate or predicted positive values)
```

```python
recall_score(y, y_pred) #sensitivity (true positive rate)
```

```python
print(classification_report(y, y_pred)) #you can compile precision and recall measures in the classification_report function
```

```python
roc_auc_score(y, y_pred) #area under the curve
```

#### Remember that the performance metrics are actually calculated by referencing a test set! The examples showed above did not split the data into test and train/validation sets

### Test-train dataset split

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
#fit logistic regression on train set
lr = LogisticRegression() #create a classifier object
lr.fit(X_train, y_train)
```

```python
y_probs_test = lr.predict_proba(X_test) #get probabilities of each class
y_pred_test = lr.predict(X_test) #predict based on test set predictors

print(y_probs_test, y_pred_test)
```

```python
confusion_matrix(y_test, y_pred_test) #calculate CM with test set reference (true classes, predicted classes)
```

```python
accuracy_score(y_test, y_pred_test) #compute the accuracy of predictions in relation to true classes
```
