## Ensembles for decision trees

Ensembles are the combination of different models to improve the predictions

You can take the average of predictions of different models and use them to compute accuracy


In the case of decision trees:

Decision trees have high variance, but low bias, and when you ensemble different number of trees the variance decreases. 

For decision trees there are three ensemble methods:

1. Bagging

2. Random Forest

3. Boosting

## Bagging

The bagging method is to "bag" up a number of different decision trees fitted in a different number of **bootstrap samples** (samples with replacement) from the same training set. So each decision tree will give a different ouput regression value or classification, and these are averaged between them all to give a final ouput estimate. 

So, 

* In regression we take the average of predicted values

* In classification we take the majority vote, most predicted class between all bagged decision trees will be taken as the final prediction


A limitation is that bagging creates correlated trees, because bagged trees will be very similar. Suppose that there are strong predictors along with modereate predictors. Most of the bagged trees will use these strong predictors to train the data and it will result in very similar bag trees. **Therefore, when you have correlated trees, averaging the decisions do not reduce the large variance of the model.**

## Random forests

Random forests algorithm is the combination of **bagging plus the randomly selection of X predictors of the dataset**. A different random subset of predictors/features is selected for each tree. This reduces the **correlation problem** caused by the bagging that creates correlated trees, thereby improving prediction accuracy.



The random forest model will fit much more smoother than the decision trees alone because it is the average of all the steps that were fitted with tree alone. ## Boosting

Process of tuning a weak learner into a strong learner

There are three process of boosting:

* Gradient Boost

* Ada Boost

* XG boost


### Gradient boost

Slow learning process. A first small tree is created.

The trees learn from residuals of trees to ajust the model. The residuals are the difference between predicted values and the actual observed values. 

The trees are then the sum of residuals from the first tree and the newly created tree multiplied by a **shrinkage parameter**.


So the trees in gradient boosting is created one at a time while fitting the residuals from the previous tree, and controlling the tree growth with the shrinkage parameter. 


When fitting the algorithm of gradient boosting, three hyperparameters have to be specified: 1) number of trees to be built; 2) shrinkage parameter lambda (which control the rate at which the model learns); 3) depth of the boosting tree, which control the growth of small trees. 






### Ada (adaptive) boosting 


The algorithm first creates a tree and predictions are made based on that tree. 

Wherever that tree has misclassified in classification tree, or wherever the residuals of a regression tree is very large, we **increase the importance of that particular misclassified observation**.

Now that we have increased the importance of that particular observation, the tree is retrained to try to capture or rightly classify those observations. 

The step of finding the misclassified or large residuals will repeat and increase the weight of those observations to run the model again. This is repeated by a number of tree times, which is a **hyperpameter** to be specified. 


This is also a form of boosting because **each tree is learning from the previous tree** 






### XG boosting


Similar to gradient boosting, the difference is that XG-boost uses regularization to control over-fitting of the model, givigin a better performance.

XG-boosting contains regularization terms in the cost function (MSE in regression)


The cost function we are trying to minimize contains a penalty term (Lasso and Ridge regression that shrinks estimates towards 0) for number of variables. We want to minimize the number of variables in final model along with the MSE or accuracy, avoiding overfitting.
