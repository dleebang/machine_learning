## Logistic regression

The logistic regression is used to limit the boundaries of the beta estimates of a regression between 0 and 1 and thus use for classification problems in trying to predict categories.

The logistic function is the same as the sigmoid function (an s-shape curve)

Looking at the graph: if X is equal to ~1500, the probability of default will be ~0.1 given a threshold (dashed line). The threshold can also be called as a *decision rule* given probabilities. The decision rule can be variable, it is 0.2 in the graph below.

Another thing to keep in mind is that we have outliers, a linear regression would change the slope and misclassify the prediction points. However, in logistic regression, as the boundaries are set to 0s and 1s, if there are outliers, the model will still capture extreme values as 0s and 1s for below and above outliers, respectively.







In linear regression model we use the ordinary least squares methods to find the beta estimates. Now, in logistic regression what is used is the **maximum likelihood method** (mxima verossimilhana).

## Interpreting results of logistic regression

The interpretation of the results will be similar to linear regression as the model gives an ouput of beta coefficients, standard errors, a z-statistics (instead of t-statistics) and a p-value.

The bottomline here is to consider a impactful/significant relationship, the p-value should be less than an established threshold of 0.05 or 0.01 (95% or 99% confidence of significant relationship). 

The H0 (null hypothesis) will consider beta coefficient of 0, while HA (alternative hypothesis) considers beta coefficient different from 0. 

## Logistic regression with multiple predictors

Now using multiple predictors to make classifications in logistic regression

We include all the combination of betas and predictors

$$p(X) = \frac{e^{\beta_{0}+\beta_{1}X_{1} + ... + \beta_{p}X_{p}}{1 + e^{\beta_{0}+\beta_{1}X_{1} + ... + \beta_{p}X_{p}}$$

**Remember**: The betas are calculate through maximum likelihood

and the boundary condition (decision rule) for probabilities of predictions to be 0 or 1 will be arbitrary due to business requirements, but we usually set it to 0.5

**Logistic regression can also be applied to multiple-class response variables, but we usually do not use it, instead apply LDA/QDA or KNN models**
