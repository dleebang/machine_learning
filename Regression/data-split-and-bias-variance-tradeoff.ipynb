## Split of data into train and test sets

For supervised machine learning algorithms, we will want to make predictions and worry about error of them into a test set, which is split from the original data set usually in a proportion of 20%.

**Most common test-train split techniques**:

* Validation set approach
    * Random division of data into two parts usually in a proportion of 80:20%
    * It is commonly used when you have large number of observaions
   
* Leave one out cross validation
    * Leave one observation every time from training set
    
* K-Fold cross validation
    * Divide the data into k sets for a number of repetitions (bootstrap samples - from edx course) and compute the mean squared error for each set, and you will end up with an average of mean squared errors across all K sets.
    * Keep one testing and K-1 for training

## Bias-variance tradeoff

Our objective is to find the model with the minimum error when applied to a **test** set

There are three contributors to the **expected test error** which is given as:

$$Expected test error = E(Bias) + E(Variance) + E(\epsilon)$$

Where $E(\epsilon)$ is the variance of error, irreducible

$E(Variance)$ is the amount by which predicted function will change if we change training dataset

$E(Bias)$ is the error due to the approximation of predictors in relation to actual observed values



As per the graph above, we see that more simple models are less flexible in capture the relationship of the points. As you increase model complexity, your model will capture well the data relationship and can achieve perfect predictions in train set as each point will be used to predict itself (minimizing the prediction error), but when you apply the model to the test set, the tendency is the error to increase and this discrepancy in error between train and test set is called as variance. Remember that this concept is tightly linked to overfitting of models (low training error and high test error).

The bias is the error when we try to fit a linear model in a dataset that have more complex relationships and its non-linear relationship could never be captured by a linear model. This is often named as the underfitting of the model. In general when you increase flexibility of the model the bias decreases. 





**The tradeoff**

If we increase model flexibility/complexity, error due to variance increases, but error due to bias decreases. Therefore the challenge is to find where the sum of the error due to bias + error of variance is minimal.
