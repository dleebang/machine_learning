#### Interpreting the Multiple linear regression

Look at the coefficients sign = if it is positive, a increase in the coefficient will cause that amount of increase in the response variable (Y). If negative, it will have the oposite relationship, for example, if we increase the value of a negative predictor, the response variable will decrease. 

Look at the significance given by the p-value and if it is less for example, than 0.05 it has a significant relationship.

Look at the confidence interval and be aware if those min and max values of the interval cross the zero.

Evaluate all these results to have a business sense of the data. In the case of the working dataset, we can say tha the price of the house is dependent on significant predictor variables such as room_num, air_qual, airport_YES, etc...

## Ridge and Lasso regression

```python
#Standardize the predictors data for ridge and lasso regression

#import the preprocessing function from sklearn
from sklearn import preprocessing
```

### Ridge

```python
scaler = preprocessing.StandardScaler().fit(X_train) #create scaler object by standardizing the raw dataset removing the mean and scaling to unit variance
```

```python
X_train_scaled = scaler.transform(X_train)
X_train_scaled 
```

```python
X_test_scaled = scaler.transform(X_test)
X_test_scaled
```

```python
#import functions from sklearn
from sklearn.linear_model import Ridge
```

```python
lm_ridge = Ridge(alpha = 0.5) #alpha = lambda
lm_ridge.fit(X_train_scaled, y_train)
```

```python
r2_score(y_test, lm_ridge.predict(X_test_scaled)) # compare the R^2 of true values in test versus the predicted values given predictors
```

```python
#test different alpha/lambda parameters
#import validation curve from sklearn
from sklearn.model_selection import validation_curve
```

```python
help(validation_curve) #call help to see the arguments of the function
```

```python
# we want to train our model to multiple values of lambda and choose the best lambda that provides the best model
lambda_range = np.logspace(-2,8,100) #logspace will create a sequence of 100 values between 10^-2 and 10^8
```

```python
lambda_range
```

```python
 #validation_curve(estimator, predictors, response, param_name, param_range, scoring)
train_scores, test_scores = validation_curve(Ridge(), X_train_scaled, y_train, "alpha", param_range = lambda_range, scoring = "r2")
```

Note that we did not provide a test set to the validation curve function. The function is performing a cross-validation of k-fold and this cross-validation already makes the calculation for test_sets.

```python
print(train_scores)
print(test_scores)
```

```python
train_mean = np.mean(train_scores, axis = 1) #take the mena values of train scores across rows
train_mean
```

```python
test_mean = np.mean(test_scores, axis = 1) #take the mena values of test scores across rows
test_mean
```

```python
max(test_mean) #get the maximum r2 of the model
```

```python
#plot to compare visually the lambda parameters with the r2 scores
sns.jointplot(x = np.log(lambda_range), y = test_mean).set_axis_labels('$\lambda$', 'Test $R^{2}$ score', fontsize=16)
```

```python
 np.where(test_mean == max(test_mean)) #find the index where the test_mean is the max
```

```python
optimal_lambda = lambda_range[30] #or you can nest the np.where code above inside the square brackets indexing
optimal_lambda
```

```python
#now fit the model with the best lambda parameter
lm_ridge_best = Ridge(alpha = optimal_lambda) #create a ridge regression with optimized lambda
lm_ridge_best.fit(X_train_scaled, y_train)
```

```python
r2_score(y_test, lm_ridge_best.predict(X_test_scaled)) #compute r2 for test with optimized lambda
```

```python
r2_score(y_train, lm_ridge_best.predict(X_train_scaled)) #compute r2 for train with optimized lambda
```

You won't see at this moment great differences in r2 depending on the model used because of small sample size of the dataset
