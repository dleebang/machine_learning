## KNN

K-Nearest Neighbor is a non-parametric algorithm, therefore we do not assume any functional form of the relationship. Therefore, the final curve of the algorithm can have a very complex relationship and can also show higher accuracies too.

The KNN makes prediction based on the average conditional probabilities of the classes within K data points neighboring that prediction point. (left graph of image below). When you have half probability of being one class or the other, the algorithm assigns the class of the predicted poitn randomly.



The right graph represents the boundary lines (decision rules) for predictions of classes. Note that the decision boundary is more flexible than a linear boundary such as that of LDA and Logistic Regression. 



The K value can be found recursevely and that with higher accuracy predictions on the **test set** can be selected. The K = 1 will use the point data itself to make prediction and (overtraining) and larger values of K can lead to oversmoothing of the prediction curve and can be unsensitive to data points. See the graph in the left below with K = 1, the decision rule/boundary becomes very complex and you are overtraining the model. The graph on the right oversmooths the decision boundary by setting K larger and do not capture the non-linear relationship of datapoints, leading to mispredictions of classifications.




K is called as the hyperparameter of the KNN classifier model. Remember about the bias-variance tradeoff here. The KNN model here is an example (depending on k values) how you can evaluate the tradeoff. If you set the k = 1, each point is being used to predict itself and then the bias of the model will be very low, but the variance will be high in the sense that in train set you have good accuracy, but when model is applied to test set the accuracy decreases. 

If you set the k = 100, the point average will be that of 100 neighbors, create high bias.


#### Important notes:

* KNN distance between observations impacts the classifier
* Scale matters, and to handle this problem of scaling, we standardize the data (mean = 0 and sd = 1)
