#### Multiple linear regression

We usually have more than one predictor and the relationship for multiple linear regression can be written as:

$$ Y = \beta_{0} + \beta_{1}X_{1} + \beta_{2}X_{2} + ... + \beta_{p}X_{p} + \epsilon $$

$\beta_{0}$ is intercept
p is the number of predictors
$\epsilon$ is error term

For the model being worked, the equation is:

Price = $\beta_{0} + \beta_{1}crimerate + \beta_{2}poorpop ... \beta_{16}avgdist$

the coefficient betas are give us the amount of increase in the price (response) variable
So for example the price will increase beta 1 units of crime rate, or beta 16 avg_dist 


#### The F-statistics

Other result that is given by the linear model output is the F-statistic

**It is important when you have several independent/predictor variables**

Remember that each independent variable will have t-value and p-value outputs

If at least one of the coefficients show a small p-value (less than the threshold of .01 or .05) it means that at least one of the predictors are related to the response variables. But when number of variables is large, there is a high chance that one or more of the betas estimates are wrongly classified as significants. 

Therefore, the F-statistic is a measure that adjust for the number of predictor variables when taking the ration of the variances:



Where p = the number of variables
TSS = total sum of squares: sum((actual value Y - mean value Y)^2)
RSS = residual sum of squares: sum((actual value Y - predict value Y)^2)
n = sample size

* If the overall p value of the F-statistics is significant, then we can look at the t and p value of individual predictors to see if there is relationship btw predictor and reponse variables. 

The F-statistics assure that you are not assuming relationship of predictor and response variables due to chance based on their coefficients and t, p -values.
