## Confusion matrix and model performance

The algorithm assigns one state (0 or 1) as positive values and calculates proportions of true/false positives and true/false negatives based on this reference.

This matrix is the proportion of positive/negative predicted values in relation to those true positive/negative. For example in the table below:



The false positive predictions are called the **type error 1** when you predicted positive but the actual value is negative

The false negative predictions are called as the **type error 2** when you predicted negative but the actual value is positive


### The model performance can be evaluated through the following measures:


* The proportions of true/false positives and true/false negatives have specific names for evaluating metrics in machine learning:

The proportion of true positives in relation to false negatives are called the **sensitivity**  (also called as recall).

The proportions of true negatives in relation to false positives are called the **specificity** (also termed as 1-False positive rate)

The specificity can also be termed as the **precision** which is *positive predicted value* (PPV), that unlike sensitivity and specifity, depends on prevalence of an event occurring and you can get higher precision of predictions even when guessing.




**Roc curve**:

The ROC (Receiver operating characteristic) curve plots sensitivity (true positive rates) versus the FPR (false positive rate, which is 1 - specificity (true negative rate)). 



We want the true positive rate curve (y axis) to be as close to values of 1 to get good classfier model performance. If the curve is close to the straight line, the model is not predicting very good and is almost similar to random classifying (guessing).
