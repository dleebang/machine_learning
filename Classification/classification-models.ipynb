## The three classification models

These models are used to predict the quality or category of the response variable instead of a true value (as in regression analysis). 

**The more common ones**:

* Logistic Regression
* Linear Discriminant Analysis
* K Nearest Neighbors

We will skip Linear discriminant analysis for now because of the petrobras concourse

#### How to evaluate the best model for your dataset

Some important questions we might seek to address in our house price dataset:

1) Prediction Question: Will the house be sold within three months of getting listed?

2) Inferential Question: How accurately can we estimate the effect of each of the predictor variables on the response variable


**So, why we cannot use linear regression to do classification?**

Linear regression cannot be used for more than two categories prediction. As explained in dummy class, you cannot create a dummy variable containing classes 0, 1, 2, and so son. Instead you have to create n-1 dummy variables containing 0s and 1s for each category. Therefore, if we want to use these dummy variables as response variables, linear regression algorithms can't handle more than one response variable. 


Another problem faced with using linear regression with classification/categorical variables is that you have resulting probability of the model that the predict value of a regression will be from -inf to +inf, but the classification response variable can only vary from 0 to 1, as the graph below. 



A third problem faced is about outlier impact. To fit extreme values below or above the distribution average of the response variable, the slope will change so the regression model can fit these outliers. Therefore, linear regression models is very sensitive to outliers, and classifiers models can handle outliers better.
